# ASSIGNMENT-1

```python
# Importing cars.csv file

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the csv file into a pandas dataframe
cars = pd.read_csv("cars-1.csv")

# print the first 5 rows of the dataframe
print(cars.head())
```

```python
print(cars.columns)
```

Generate a line plot graph for the column ‘model’ and ‘hp’.

a. Map the ‘model’ column on the x-axis.

b. Map the ‘hp’ column on the y-axis.

c. Provide the x-axis label as Models of the cars.

d. Provide the y-axis label as Horse-Power of Cars.

e. Set the title as Model Names vs HorsePower.

```python
# Create the line plot
plt.plot(cars["model"], cars["hp"])

# Set the x-axis label
plt.xlabel('Models of the cars')

# Set the y-axis label
plt.ylabel('Horse-Power of Cars')

# Set the title
plt.title('Model Names vs HorsePower')

# Show the plot
plt.show()
```

2. Write code to load data from cars and print a bar graph of count of columns with null values.

```python
# Count null values in each column
null_counts = cars.isnull().sum()

# Plot bar graph of null counts
null_counts.plot(kind='bar')
# Plot bar graph
plt.bar(null_counts.index, null_counts.values)
plt.xlabel('Columns')
plt.ylabel('Number of null values')
plt.title('Count of null values in each column')
plt.show()
```
